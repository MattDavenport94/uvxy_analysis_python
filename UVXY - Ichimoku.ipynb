```python
import datetime
```

```python
start = datetime.datetime(2017,4,24)
end = datetime.datetime(2019,4,24)
```

```python
UVXY = web.DataReader('UVXY','yahoo',start,end)
```

```python
UVXY = pd.DataFrame(UVXY)
```

```python
UVXY = UVXY.reset_index()
UVXY['Date'] = pd.to_datetime(UVXY['Date'],format='%Y-%m-%d')
```

```python
UVXY['tenkan_sen'] = (UVXY['High'].rolling(window = 9).max() +
UVXY['Low'].rolling(window = 9).min()) /2 #Tenkan-sen Line

UVXY['kijun_sen'] = (UVXY['High'].rolling(window=26).max() + 
UVXY['Low'].rolling(window=26).min()) /2 #Kijun-sen Line

UVXY['chikou_span'] = (UVXY['Close'].shift(-26))
```

```python
last_date = UVXY['Date'].iloc[-1]

for i in range(26):
    UVXY = UVXY.append(
        {"Date":last_date + datetime.timedelta(days=i+1)}, 
        ignore_index=True)

```

```python
UVXY['senkou_span_a'] = ((UVXY['tenkan_sen'] + UVXY['kijun_sen']) 
                             /2).shift(26)

high_52 = UVXY['High'].rolling(window=52).max()
low_52 = UVXY['Low'].rolling(window=52).min()

UVXY['senkou_span_b'] = ((high_52 + low_52) /2).shift(26)
```

```python
#Formulas for Fractals are
#Brearish Fractal = High(N)>High(N-2) and 
#                   High(N)>High(N-1) and
#                   High(N)>High(N+1) and
#                   High(N)>High(N+2)

#Bullish Fractal = Low(N)<Low(N-2) and
#                  Low(N)<Low(N-1) and
#                  Low(N)<Low(N+1) and
#                  Low(N)<Low(N+2)

# N = High/low of the current price bar
# N-2 = High/low of price bar two periods to the left of N
# N-1 = High/low of price bar one period to the left of N
# N+1 = High/low of price bar one period to the right of N
# N+2 = High/low of price bar two periods to the right of N
```
